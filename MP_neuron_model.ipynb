```python
# AND Gate
import numpy as np
# Step 1: Define inputs as NumPy arrays
x1 = np.array([0, 0, 1, 1])
x2 = np.array([0, 1, 0, 1])

# Step 2: Define weights and threshold
w1 = 1
w2 = 1
threshold = 2

# Step 3: Define threshold function
def mp_threshold(net_input, threshold):
    return 1 if net_input >= threshold else 0

# Step 4: Calculate and display output
print("x1 x2  w1 w2  t  O")
for i in range(len(x1)):
    net = x1[i]*w1 + x2[i]*w2   # weighted sum
    output = mp_threshold(net, threshold)
    print(f"{x1[i]}  {x2[i]}   {w1}  {w2}  {threshold}  {output}")

```

```python
#XOR Gate
import numpy as np

# Inputs
x1 = np.array([0, 0, 1, 1])
x2 = np.array([0, 1, 0, 1])

# Threshold value (only one)
threshold = 1

# Threshold function
def mp_threshold(net, t):
    return 1 if net >= t else 0

print("x1 x2  w1  w2  t  O")

for i in range(len(x1)):
    # Assign w1 and w2 according to your pattern
    if   (x1[i], x2[i]) == (0,0): w1, w2 =  1,  1
    elif (x1[i], x2[i]) == (0,1): w1, w2 = -1,  1
    elif (x1[i], x2[i]) == (1,0): w1, w2 = -1, -1
    elif (x1[i], x2[i]) == (1,1): w1, w2 =  1, -1

    # Calculate net input
    net = x1[i]*w1 + x2[i]*w2

    # Apply single threshold
    output = mp_threshold(net, threshold)

    print(f"{x1[i]}  {x2[i]}   {w1}  {w2}  {threshold}  {output}")
```
